```python
import pandas as pd

from datasets import Dataset
```

```python
data_head = pd.read_csv("D:\Gen AI\Generative AI\materials\Assignment\Resume\Resume.csv")
```

```python
data_head
```

```python
data_head["Category"][0]
```

```python
# Create a dictionary for each column
data_dict = {
    "id": data_head["ID"].tolist(),
    "resume_text": data_head["Resume_str"].tolist(),
    "html_data": data_head["Resume_html"].tolist(),
    "category": data_head["Category"].tolist(),
}
```

```python
# Explicitly convert columns to strings
data_head["ID"] = data_head["ID"].astype(str)
data_head["Resume_str"] = data_head["Resume_str"].astype(str)
data_head["Resume_html"] = data_head["Resume_html"].astype(str)
data_head["Category"] = data_head["Category"].astype(str)

# Create a Hugging Face Dataset
data_dict = {
    "id": data_head["ID"].tolist(),
    "resume_text": data_head["Resume_str"].tolist(),
    "html_data": data_head["Resume_html"].tolist(),
    "category": data_head["Category"].tolist(),
}

dataset = Dataset.from_dict(data_dict)
```

```python

# Save the dataset to a local directory using forward slashes
dataset.save_to_disk("D:/Gen AI/Generative AI/materials/Assignment/Resume/resume_huggingface_dataset")
```
