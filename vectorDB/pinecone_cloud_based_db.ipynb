## Extract the Text from the PDF's

```python
from langchain.document_loaders import PyPDFDirectoryLoader  # to extract the dircetory contained with pdf files
```

```python
!pip install --upgrade PyPDF2
```

```python
# loading the pdfs directory
loader = PyPDFDirectoryLoader("pdfs")
data = loader.load()
```

```python
data
```

## Chunkins

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
text_chunks = text_splitter.split_documents(data)
```

```python
text_chunks
```

```python
len(text_chunks)
```

```python
text_chunks[1]
```

## Embeddings

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
!pip install sentence-transformers
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
query_result = embeddings.embed_query("Hello World")
```

```python
query_result
```

```python
print("Length", len(query_result))
```

## Initializing the Pinecone

```python
import os
```

```python
PINECONE_API_KEY = os.environ.get('PINECONE_API_KEY', 'bd47093c-0a07-4982-8904-ae234ec244c5')
PINECONE_API_ENV = os.environ.get('PINECONE_API_ENV', 'gcp-starter')
```

```python
import pinecone
# initialize pinecone
pinecone.init(
    api_key=PINECONE_API_KEY,  # find at app.pinecone.io
    environment=PINECONE_API_ENV  # next to api key in console
)
index_name = "test" # put in the name of your pinecone index here
```

```python
from langchain.vectorstores import Pinecone
```

```python
docsearch = Pinecone.from_texts([t.page_content for t in text_chunks], embeddings, index_name=index_name)
```

## If you already have an index, you can load it like this

```python
docsearch = Pinecone.from_existing_index(index_name, embeddings)
docsearch
```

## Similarity Search

```python
query = "What is yolo?"
```

```python
docs = docsearch.similarity_search(query, k=3)
```
